```python
## Operadores Lógicos em Python
Vamos começar entendendo os operadores básicos.
```

```python
# Operadores lógicos básicos

p = True
q = False

print("p =", p)
print("q =", q)

print("p and q =", p and q)  # AND
print("p or q =", p or q)    # OR
print("not p =", not p)      # NOT
```

```python
## 1ª Lei de De Morgan

A negação de uma conjunção:

¬(p ∧ q) = ¬p ∨ ¬q
```

```python
# 1ª Lei de De Morgan

p = True
q = False

esquerda = not (p and q)
direita = (not p) or (not q)

print("¬(p ∧ q) =", esquerda)
print("¬p ∨ ¬q =", direita)
```

```python
## 2ª Lei de De Morgan

A negação de uma disjunção:

¬(p ∨ q) = ¬p ∧ ¬q
```

```python
# 2ª Lei de De Morgan

p = True
q = False

esquerda = not (p or q)
direita = (not p) and (not q)

print("¬(p ∨ q) =", esquerda)
print("¬p ∧ ¬q =", direita)
```

```python
## Tabela Verdade

Vamos verificar automaticamente se as leis são válidas.
```

```python
# Tabela verdade automática

valores = [True, False]

print("p\tq\t¬(p∧q)\t¬p∨¬q")

for p in valores:
    for q in valores:
        esquerda = not (p and q)
        direita = (not p) or (not q)
        print(p, q, esquerda, direita, sep="\t")
```

```python
## Validação das Leis

Vamos testar se as leis são sempre verdadeiras.
```

```python
# Validador das Leis de De Morgan

valores = [True, False]
valido = True

for p in valores:
    for q in valores:
        if not (not (p and q) == ((not p) or (not q))):
            valido = False

        if not (not (p or q) == ((not p) and (not q))):
            valido = False

if valido:
    print("✔ As Leis de De Morgan são sempre verdadeiras!")
else:
    print("✘ Existe erro nas leis")
```

```python
## Aplicação prática em programação

Exemplo de sistema de acesso.
```

```python
# Aplicação prática

usuario_logado = False
tem_permissao = False

# Forma normal
if not (usuario_logado and tem_permissao):
    print("Acesso negado")

# Usando De Morgan
if (not usuario_logado) or (not tem_permissao):
    print("Acesso negado (forma equivalente)")
```
